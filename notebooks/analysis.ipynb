# Healthcare Survey Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/survey_data.csv')
df.head()
```

```python
income_by_age = df.groupby('age')['income'].mean()
income_by_age.plot(kind='bar')
plt.title('Average Income by Age')
plt.xlabel('Age')
plt.ylabel('Income')
plt.tight_layout()
plt.savefig('../charts/age_income.png')
plt.show()
```

```python
spending = df.groupby('gender')[['utilities','entertainment','schoolfees','shopping','healthcare']].sum()
spending.plot(kind='bar')
plt.title('Gender Distribution Across Spending Categories')
plt.ylabel('Amount Spent')
plt.tight_layout()
plt.savefig('../charts/gender_spending.png')
plt.show()
```
